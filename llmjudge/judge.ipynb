```python
from openai import OpenAI
from dotenv import load_dotenv
from collections import Counter
import random
```

```python
load_dotenv()
client = OpenAI(api_key='your-api-key-here')

random.seed(1)  # sets the seed to 1

random_numbers = random.sample(range(1, len(results['trained_output'])), 100)
print(random_numbers)
```

```python
responses = []
reverse = True
for n in random_numbers:
  if reverse:
      question = results['prompt'][0]
      answer_a = results['rejected'][0]
      answer_b = results['trained_output'][0]
      prompt = prompt_template.format(instructions = system_prompt, question = question, answer_a = answer_a, answer_b = answer_b)

  else:
      question = results['prompt'][0]
      answer_a = results['trained_output'][0]
      answer_b = results['rejected'][0]
      prompt = prompt_template.format(instructions = system_prompt, question = question, answer_a = answer_a, answer_b = answer_b)

  print(prompt)
  completion = client.chat.completions.create(
    model="gpt-4-turbo",
    messages=[
      {"role": "system", "content": system_prompt},
      {"role": "user", "content": prompt}
    ]
  )
  responses.append(completion.choices[0].message)
```

```python
c = Counter()
for r in responses:
    content = r.content
    if "[[A]]" in content:
        c['A'] += 1
    elif '[[B]]' in content:
        c['B'] += 1
    elif '[[C]]' in content:
        c['C'] += 1
```
